```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel("UFCFighters.xlsx")
```

```python
data.head()
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.groupby("Country").count()["Name"].plot()
```

```python
data["Wins"]=data["Record (W-L-D)"].apply(lambda x: str(x).split("-")[0])
```

```python
data["Wins"]=pd.to_numeric(data["Wins"],errors="ignore")
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python

data["Wins"]=pd.to_numeric(data["Wins"],errors="ignore")
```

```python
data.hist()
```

```python
data["Loss"]=data["Record (W-L-D)"].apply(lambda x: str(x).split("-")[1])
data["Loss"]=pd.to_numeric(data["Loss"],errors="ignore")
```

```python
data["Draw"]=data["Record (W-L-D)"].apply(lambda x: str(x).split("-")[2])
data["Draw"]=pd.to_numeric(data["Draw"],errors="ignore")
```

```python
data.hist(figsize=(15,15))
```

```python
data.groupby("Country").sum()["Wins"].plot.bar(figsize=(20,10))
```

```python
data.groupby("Country").sum()["Loss"].plot.bar(figsize=(20,10))
```

```python
data.groupby("Country").sum()["Draw"].plot.bar(figsize=(20,10))
```
